```python
df = df_filt
```

```python
df.drop(['Div', 'Time'], axis = 1, inplace = True)
```

```python
df.head(5)
```

```python
df.drop(df.columns[40:], axis = 1, inplace = True)
```

```python
df['AvgBH'] = df[['B365H', 'BWH', 'IWH', 'PSH', 'WHH', 'VCH']].mean(axis=1)
df['AvgBD'] = df[['B365D', 'BWD', 'IWD', 'PSD', 'WHD', 'VCD']].mean(axis=1)
df['AvgBA'] = df[['B365A', 'BWA', 'IWA', 'PSA', 'WHA', 'VCA']].mean(axis=1)
```

```python
df.drop(df.columns[22:40], axis = 1, inplace = True)
```

```python
df['FTRn'] = df.FTHG - df.FTAG
```

```python
df.Date = df.Date.apply(lambda x: int(str(x)[3:5]))
```

```python
df.drop(['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HTR', 'Referee'], axis = 1, inplace = True)
```

```python
df = df[['Date', 'HomeTeam', 'AwayTeam', 'AvgBH', 'AvgBD', 'AvgBA', 'FTRn', 'FTR']]
```

```python
df.to_csv('csvs/premodel0.csv', index = False)
```

```python
df = pd.concat(map(pd.read_csv, ['csvs/premodel.csv', 'csvs/premodel0.csv']), ignore_index = True)
```

```python
y = df['FTRn']
x = df.drop(['FTRn', 'FTR'], axis = 1)
```

```python
dummy = pd.get_dummies(df['HomeTeam'])
x = pd.concat([x, dummy], axis = 1)
dummy = pd.get_dummies(df['AwayTeam'])
x = pd.concat([x, dummy], axis = 1)
```

```python
x.drop(['HomeTeam', 'AwayTeam'], axis = 1, inplace = True)
```

```python
x = x.tail(27)
```

```python
x.shape
```

```python
y_pred_svr = svr_model.predict(x)
```

```python
y_pred_en = en_model.predict(x)
```

```python
df = pd.read_csv('csvs/premodel0.csv')
```

```python
df['FTRn SVR'] = y_pred_svr
df['FTRn EN'] = y_pred_en
```

```python
df['FTRn SVR'] = df['FTRn SVR'].round(decimals = 2)
df['FTRn EN'] = df['FTRn EN'].round(decimals = 2)
```

```python
df
```

```python
df['FTRnp'] = (df['FTRn EN'] + df['FTRn SVR'])/2 
```

```python
df.head()
```

```python
def resultp(res):
    if((res >= -0.5) & (res <= 0.5)):
        return 'D'
    elif(res > 0.5):
        return 'H'
    elif(res < -0.5):
        return 'A'
```

```python
df['FTRp1'] = df['FTRnp'].apply(resultp)
```

```python
df
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(df['FTR'], df['FTRp1'])
```
