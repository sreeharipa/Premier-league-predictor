```python
def resultp(res):
    if((res >= -0.5) & (res <= 0.5)):
        return 'D'
    elif(res > 0.5):
        return 'H'
    elif(res < -0.5):
        return 'A'
```

```python
df['FTRp1'] = df['FTRnp'].apply(resultp)
```

```python
for i, j, k, l in df[['HomeTeam', 'AwayTeam', 'FTRp1', 'FTRnp']].itertuples(index = False):
    table.loc[(table['Team'] == i) & (k == 'H'), 'Won'] = table.Won + 1
    table.loc[(table['Team'] == j) & (k == 'A'), 'Won'] = table.Won + 1
    table.loc[((table['Team'] == i) | (table['Team'] == j)) & (k == 'D'), 'Drawn'] = table.Drawn + 1
    table.loc[(table['Team'] == i) & (k == 'H'), 'GDn'] = table.GDn + l
    table.loc[(table['Team'] == j) & (k == 'A'), 'GDn'] = table.GDn - l
    table.loc[(table['Team'] == i) & (k == 'A'), 'GDn'] = table.GDn + l
    table.loc[(table['Team'] == j) & (k == 'H'), 'GDn'] = table.GDn - l

table.Lost = table.Played - (table.Won + table.Drawn)
table.Points = 3*table.Won + table.Drawn
table = table.sort_values(['Points', 'GDn'], ascending = False)    
```
