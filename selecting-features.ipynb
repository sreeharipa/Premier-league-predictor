```python
df['AvgBH'] = df[['B365H', 'BWH', 'IWH', 'PSH', 'WHH', 'VCH']].mean(axis=1)
df['AvgBD'] = df[['B365D', 'BWD', 'IWD', 'PSD', 'WHD', 'VCD']].mean(axis=1)
df['AvgBA'] = df[['B365A', 'BWA', 'IWA', 'PSA', 'WHA', 'VCA']].mean(axis=1)

```

```python
df.head(1)
```

```python
df.drop(df.columns[22:40], axis = 1, inplace = True)
```

```python
df.head(1)
```

```python
df.Referee.value_counts()
```

```python
df['FTRn'] = df.FTHG - df.FTAG
```

```python
df.head(1)
```

```python
df.Date = df.Date.apply(lambda x: int(str(x)[3:5]))
```

```python
df.sample(10)
```

```python
df.drop(['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HTR', 'Referee'], axis = 1, inplace = True)
```

```python
df = df[['Date', 'HomeTeam', 'AwayTeam', 'AvgBH', 'AvgBD', 'AvgBA', 'FTRn', 'FTR']]
```
